```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
loan = pd.read_csv('./lending-club-loan-data/loan.csv', low_memory=False)
```

```python
print ('# or rows: {}, # of columns: {}'.format(loan.shape[0], loan.shape[1]))
loan.head()
```

```python
loan.info()
#there are a number of columns that have a lot of nulls since the largest number of entries in a column is 887,379 records while
#columns at the end are around 21,372 and the joint columns have only around 500 records.
```

```python
print(loan.columns)
```

```python
pd.set_option('display.max_columns', None)
loan.describe()
# loan_amnt, funded_amnt, and funded_amnt_inv have pretty close stats
# dti with a max 9999 compared to the 75% range of 23.95 means there are very large outliers
```

```python
loan[loan.dti>100]
# 11 records out of 887379 records with a dti value greater than 100, might want to drop the large outliers
```

```python
loan[loan.annual_inc < 1000]
# the two records that have a dti of 9999.00 also have a annual income of zero, the two important factors that determine if
# borrower can repay loan in full is missing or incomplete so it is better to drop these two records
```

```python
loan.drop(484446, inplace=True)
loan.drop(531886, inplace=True)
loan[loan.annual_inc == 0]
```

```python
#loan amount applied and the amount funded has close stats and distribution, meaning all loans applied have been funded
fig, ax = plt.subplots(1, 3, figsize=(16,5))
sns.distplot(loan.loan_amnt, ax=ax[0]);
sns.distplot(loan.funded_amnt, ax=ax[1]);
sns.distplot(loan.funded_amnt_inv, ax=ax[2]);
```

```python
print(loan.application_type.value_counts())
#number of joint applications is only around 500 records vs. 886 thousand individual

print(loan.loan_status.value_counts())
#Fully Paid and Charged off has a 2nd classification, "Does not meet the credit policy". These can be grouped into Fully Paid or Charged off
```

```python
# grouping Does not meet credit policy into Fully Paid and Charged off
loan.loan_status.replace(['Does not meet the credit policy. Status:Fully Paid'], 'Fully Paid', inplace=True)
loan.loan_status.replace(['Does not meet the credit policy. Status:Charged Off'], 'Charged Off', inplace=True)
loan.groupby('application_type').loan_status.value_counts()
```

```python
# number of borrowers that own, rent, or have a mortgage; majority either carry a mortgage or is renting
print(loan.groupby('application_type').home_ownership.value_counts())

# the purpose of the loan; majority is borrowing to consolidate debt or to cover credit card balances
print(loan.groupby('application_type').purpose.value_counts())

# there aren't much variance between individual and joint application types in each category
```

```python
loan.groupby('grade').sub_grade.value_counts().plot(kind='barh', figsize=(20,20));
#the loans are mainly concentrated in grades B and C
```

```python
loan.boxplot(column='int_rate', by='grade', figsize=(15,8), sym='k');
plt.xlabel('Grade');
plt.ylabel('Interest Rate');
# Higher grade has lower interest rate while lower grade has higher interest rates but there are some outliers where lower grade
# has some lower interest rates as well
```

```python
loan.plot(kind='scatter', x='int_rate', y='loan_amnt');
# there seems to be no relationship between interest rate and loan amount so the amount on the loan does not affect interest rate
```

```python
loan.boxplot(column='int_rate', by='loan_status', figsize=(15,8), sym='k');
plt.xlabel('Loan Status');
plt.ylabel('Interest Rate');
# Current, Fully Paid, and Issued has the lower interest rate mean/range while the loans that are in delinquency have higher/range
# though there are some outliers on the loans that are considered to be in good status
```

```python
loan.boxplot(column='dti', by='loan_status', figsize=(15,8), sym='k', showfliers=False);
plt.xlabel('Loan Status');
plt.ylabel('Debt to Income Ratio');
# lower debt to income ratio is more likely to repay loan in full and higher is more likely to default
```

```python
# average income per state and average loan amount
loan.groupby('addr_state')['annual_inc', 'loan_amnt'].mean().sort_values(by = 'loan_amnt', ascending = False)
```

```python
loan['issue_mo'], loan['issue_yr'] = loan['issue_d'].str.split('-', 1).str
fig, ax = plt.subplots(1, 2, figsize=(16,5));
loan.issue_yr.value_counts().plot(kind='bar', ax=ax[0]);
loan.issue_mo.value_counts().plot(kind='bar', ax=ax[1]);
# most of the loan are initiated in 2014 and 2015
# more people are initiating the loan during the later half of the year
```

```python
# state with more large # of accounts in delinquent
loan.groupby('addr_state').acc_now_delinq.agg(['max', 'mean']).sort_values(by=['max', 'mean'], ascending=False)
```

```python
#assigning sub_grade to numbers so we can use it in correlation
print(loan.sub_grade.unique())
loan['subgrade'] = loan.sub_grade.map({'A1':5,'A2':10,'A3':15,'A4':20,'A5':25,\
                                       'B1':30,'B2':35,'B3':40,'B4':45,'B5':50,\
                                      'C1':55,'C2':60,'C3':65,'C4':70,'C5':75,\
                                      'D1':80,'D2':85,'D3':90,'D4':95,'D5':100,\
                                      'E1':105,'E2':110,'E3':115,'E4':120,'E5':125,\
                                      'F1':130,'F2':135,'F3':140,'F4':145,'F5':150,\
                                      'G1':155,'G2':160,'G3':165,'G4':170,'G5':175})
loan.subgrade.value_counts()
```

```python
#assigning loan_status to numbers so we can use it in correlation
#large amount of data is concentrated in 'Currrent' status which does not tell us much of the possibility
#of default vs. paying in full, it is better to exclude the Current, Issued, and In Grace Period data rows
print(loan.loan_status.unique())
loan['status'] = loan.loan_status.map({'Fully Paid':0,'Charged Off':6,'Current':3,'Default':7,\
                                      'Late (31-120 days)':5,'In Grace Period':2,'Late (16-30 days)':4,\
                                      'Issued':1})
loan.status.value_counts()
```

```python
#correlation of status and subgrade with the float variables
#status has higher negative correlation with the payment that has been made and positive correlation
#with outstanding balance, it also has positive correlation with recoveries, subgrade, and dti
#subgrade has high correlation (0.98) with interest rate, which means the interest rate is probably determined by the grade assigned by LC
#subgrade also have higher correlation with balance to credit limit, inquiries, and dti, and loan amount
plt.figure(figsize=(15,30))
sns.heatmap(loan.corr()[['status', 'subgrade']], annot=True);
```
